Se todas as linhas fossem de um mesmo ativo, era só fazer

```python
df['Soma Ponderada (R$)'] = df['Quantidade'] * df['Compra (R$)']

SomaPesos = df['Quantidade'].sum()

SomaTotal = df['Soma Ponderada (R$)'].sum()

media = SomaTotal/SomaPesos

df.drop(columns = 'Soma Ponderada', inplace = True)

df
# print(media)
```

Agora que vimos como criar um conjunto contendo valores únicos de uma lista, vamos criar sub DataFrames da tabela original

```python
conjunto = set(df['Ativo'])

for nome in conjunto:
    print(df.loc[df['Ativo'] == nome] )
    print()
```

Calculando o preço médio em cada sub DataFrame

```python
from pandas import DataFrame

ListaAtivos = []
ListaPrecoMedio = []

for nome in conjunto:

    ListaAtivos.append(nome)
    
    df2 = df.loc[ df['Ativo'] == nome ]
    
    SomaPesos = df2['Quantidade'].sum()

    SomaTotal = df2['Soma Ponderada (R$)'].sum()

    media = SomaTotal/SomaPesos
    
    ListaPrecoMedio.append(media)

dicionario = {'Ativo': ListaAtivos, 'Preço Médio de compra (R$)': ListaPrecoMedio}

df3 = DataFrame(dicionario)

df3
```

```python
df3.to_csv('PrecosMedios.csv', index = False)
```
